In many situations differencing converts nonstationary series to stationary.

These nonstationary series can be thought of being composed of a nonstationary component and a zero-mean stationary component.

Example: Random walk: 

$x_t=x_{t-1}+w_t\implies \nabla x_t = w_t$,


$x_t=\beta_0+\beta_1 t+y_t\implies \nabla x_t=\beta_1+\nabla y_t$ where $y_t$ is stationary.


The integrated ARMA (ARIMA)model is a broadening of the class of ARMA models to include differencing.

## Definition (ARIMA(p,d,q))

A process $x_t$ is said to be ARIMA(p,d,q) if the d times differenced series is ARMA(p,q).

## Building ARIMA models

Basic steps

1) plot data
2) transform data if needed
3) identify the dependence orders of the model
* Be careful not to overdifference because this may introduce dependence where none exists.
* A slow decay in sample ACF $\hat{\rho}(h)$ is an indication that differencing may be needed.
* Look at the sample ACF and PACF after differencing.
* Follow the summary table in the previous section to choose preliminary values of p and q (note that it cannot be the case that both ACF and PACF tails off).
4) estimate parameters
5) diagnostic
6) model choice

```python
fig, axes = plt.subplots(nrows=3, figsize=(7, 6))
df["rate"].plot(ax=axes[0], title="Recruitment", legend=False)
sm.graphics.tsa.plot_acf(df["rate"][1:], lags=48, ax = axes[1], title="ACF")
sm.graphics.tsa.plot_pacf(df["rate"][1:], lags=48, ax = axes[2], title="PACF")
fig.tight_layout()
plt.show()
```

We might feel that the ACF is cutting off at lag 2 and PACF is tailing off. 

This suggests that the GNP growth rate is an MA(2) process (or log GNP is a ARIMA(0,1,2))

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model=ARIMA(df["rate"][1:],order=(0,0,2))
model_fit=model.fit()
```

```python
model_fit.summary()
```

Model:

$$\hat{r}_t = 0.0083+0.3067\hat{w}_t+0.2249\hat{w}_{t-1}+\hat{w}_t$$

where $\hat{\sigma}_w = \sqrt{8.92*10^{-5}} = 0.0094$ with 222-3 = 219 degree of freedom. Note that all the coefficients are significant.

```python
fig = model_fit.plot_diagnostics(figsize=(10, 7))
fig.tight_layout()
plt.show()
```

## Diagnostic

Top left - standardized residuals, Top right - histogram of standardized residuals, Bottom left - QQ-plot of residuals
and Bottom right - ACF of residuals.

Standardized residuals plot shows not obvious patterns. May be outliers, few values exceeding 3 standard deviations.

ACF shows no apparent departure from model assumptions and values are not significant for any of the shown lags.

QQ-plot (also histogram) shows that the assumption of normality is reasonable, except for some possible outliers.

## Another model to try:

Based on the ACF and PACF plots on the growth rate data. We also try AR(1) model
(we suspect that ACF tails off and PACF cuts off at lag 1).



```python
model=ARIMA(df["rate"][1:],order=(1,0,0))
model_fit=model.fit()
```

```python
model_fit.summary()
```

We see that all the coefficients are significant.

```python
fig = model_fit.plot_diagnostics(figsize=(10, 7))
fig.tight_layout()
plt.show()
```

Perhaps combining both models, i.e. ARMA(1,2) on the growth rate would be the best.

Note that we have to be concerned with overfitting, it is not always the case that more is better.
